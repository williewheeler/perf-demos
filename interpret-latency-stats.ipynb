# Interpreting Latency Statistics

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Latency model

```python
MU = 3.8
SIGMA = 1.8
```

```python
def plot_lognormal_pdf(mu, sigma):
    x = np.linspace(0.1, 600, 1_000)
    pdf = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((np.log(x) - mu) / sigma)**2)
    
    plt.plot(x, pdf, label='Lognormal PDF')
    plt.title('Lognormal Probability Density Function')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()
```

```python
plot_lognormal_pdf(MU, SIGMA)
```

```python
rng = np.random.default_rng()
```

```python
def generate_sample_stats(mu, sigma, size, num_samples):
    # TODO Treat these as time series and plot them along with the dist stats.
    # Don't need to plot all of them, but just enough to give an idea.
    result = {'mean': [], 'p2': [], 'p99': []}
    # E.g., iterate over a number of days
    for i in range(num_samples):
        # E.g., request data for a single day
        data = rng.lognormal(mean=mu, sigma=sigma, size=size)
        result['mean'].append(np.mean(data))
        result['p2'].append(np.quantile(data, 0.02))
        result['p99'].append(np.quantile(data, 0.99))
    return result
```

```python
result = generate_sample_stats(MU, SIGMA, 10_000, 10_000)
```

```python
plt.hist(result['mean'], bins=31)
plt.show()
```

```python
plt.hist(result['p2'], bins=31)
plt.show()
```

```python
plt.hist(result['p99'], bins=31)
plt.show()
```
